#### **Introduction**

<p>
Pokémon is a video game franchise centered around fictional creatures called 'Pokémon' which humans catch, train, and battle for sport. In the game, the player is tasked with building a team of strong Pokémon which they will use in to challenge other 'Pokemon Trainers'. Most games follow a linear progression in which the player travels from town to town battling powerful trainers known as 'Gym Leaders' that the trainer must defeat in order to collect a gym badge. Once all badges have been collected the player can challenge the Elite Four, four of the most powerful pokemon trainers in the reigion, which the player can battle in succession in order to reach the Regional champion. Defeating the Regional champion will crown the player as the new champion. 
</p>

<p>
Once the player becomes the new champion, they can usually engage in a quest which will revolve around tracking down 'Legendary' pokemon. Legendary pokemon are incredibly rare, and are usually very powerful.
</p>

<p>
The central game mechanic of Pokémon games is the turn-based battle system. Players take turns preforming one of three actions: preforming a move, using an item, or switching out a pokemon with another Pokemon in their party. The objective is to lower the HP of all your opponents Pokémon to 0(causing them to faint). Pokemon can lower an opponents HP by preforming moves, which will apply damage based on the attacker's Atk. or Sp. Atk. stats and the defenders Def. or Sp. Def. stats. The type of stat is determinted by the move type (either physical or special). The Speed statistic determines which Pokemon moves first in battle. The Pokemon with the higher speed statistic will always move first. Pokemon battles also make use of a rock-paper-scissors esque type system. Types are elemental attributes that determine a Pokemon's strengths and weaknesses. Pokémon take double damage from attacking moves of types they are weak to and half damage from moves they resist. For more details check out this 
<a href="https://raw.githubusercontent.com/mgfalzon/Final-Tutorial/main/type_chart.png">Type Chart</a>. 

</p>

In order to preform our analysis we'll be looking at 3 characterstics, a pokemon's type, their base stats, and finally their legendary status. We want to know if we can use these characterstics in order to predict the outcome of future battles. #### **Tidy Data: Combats**

Explanation

```python
combats.head()
```

```python
# Insert Loser column
combats['Loser'] = pd.Series(np.where(combats['Winner'] == combats['First_pokemon'], combats['Second_pokemon'], combats['First_pokemon']))
combats.head(3)
```

```python
# Find missing data
combats.isna().sum()
```

```python
# Generate win and loss counts for each pokemon
wins = combats['Winner'].value_counts().sort_index().rename('Wins')
loss = combats['Loser'].value_counts().sort_index().rename('Loss')

# Verify that each pokemon has at least one loss and one win
display(f"Total Pokemon: {len(pokemon)}")
display(f"Unqiue Winners: {len(wins)}")
display(f"Unqiue Losers: {len(loss)}")
```

17 Pokemon do not appear in the winner column and 16 Pokemon do not appear in the loser column. Let's check to see if any of these Pokemon don't appear in either column.

```python
# Pokemon that did not win or lose (these pokemon did not compete)
ids = [x for x in pokemon['#'] if x not in wins.index and x not in loss.index]
display(ids, pokemon[pokemon['#'].isin(ids)].head())
```

```python
# Let's drop those pokemon, they won't provide insight for our analysis
pokemon = pokemon[~pokemon['#'].isin(ids)]
```

```python
# Let's check again to see if each pokemon has at least one loss and one win
display(f"Total Pokemon: {len(pokemon)}")
display(f"Unqiue Winners: {len(wins)}")
display(f"Unqiue Losers: {len(loss)}")
```

There appears to be one pokemon that did not win at all. Let's see who it is

```python
# Pokemon with no wins
worst_pokemon = [x for x in pokemon['#'] if x in loss.index and x not in wins.index][0]
pokemon[pokemon['#'] == worst_pokemon]
```

```python
# Add 0 to wins for Shuckle
wins[worst_pokemon] = 0
```

```python
# Final verification
display(f"Total Pokemon: {len(pokemon)}")
display(f"Unqiue Winners: {len(wins)}")
display(f"Unqiue Losers: {len(loss)}")
```

#### **Data Exploration**

```python
# Calculate win percentage
res = pd.concat([wins, loss], axis=1)
res['win_loss'] = res['Wins'] / (res['Wins'] + res['Loss'])
res['win_pct']  = (res['win_loss'] * 100).round(1)
res.head()
```

```python
# Join with pokemon table
pokemon = pokemon.join(res, on='#')
```

```python
# Top 50 Pokemon by win percentage
top50 = pokemon.sort_values(by='win_pct', ascending=False).head(50)
top50[['Name', 'win_pct']].head()
```

```python
# Bottom 50 Pokemon by win percentage
bot50 = pokemon.sort_values(by='win_pct', ascending=True).head(50)
bot50[['Name', 'win_pct']].head()
```

```python
# Type frequency
type_freq = pokemon.groupby(by=['Type 1', 'Type 2'], dropna=False)['Name'].count().rename('freq')
type_freq = type_freq.sort_values(ascending=False)
pd.DataFrame(type_freq).head()
```

```python
# Top 50 pokemon Type frequency
type_freq = top50.groupby(by=['Type 1', 'Type 2'], dropna=False)['Name'].count().rename('count').reset_index()
type_freq = type_freq.sort_values(by='count', ascending=False).reset_index(drop=True)
type_freq.head()
```

```python
# Bottom 50 pokemon Type frequency
type_freq = bot50.groupby(by=['Type 1', 'Type 2'], dropna=False)['Name'].count().rename('count').reset_index()
type_freq = type_freq.sort_values(by='count', ascending=False).reset_index(drop=True)
type_freq.head()
```
